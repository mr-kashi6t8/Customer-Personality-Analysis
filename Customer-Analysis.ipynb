# What Are the Most frequently Purchased Products 

```python
data.columns
```

```python
Products = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds']
total_sum = data[Products].sum()
total_sum
```

```python
plt.figure(figsize=(10,6))
sns.barplot(total_sum,color='salmon',width=0.5)
plt.title('Total Spending On Products',weight='bold')
plt.xlabel('Products')
plt.ylabel('Total Spending')
```

# How does the rate of spending on product differ between different age Group ?

```python
data['age'] = 2024 - data['Year_Birth']
# data['age']
age_bins = [20,30,40,50,60,70,80,90]
data['Age_Group']=pd.cut(data['age'],bins=age_bins)
# data['Age_Group']
spending_By_AgeGroup =data.groupby("Age_Group")[Products].mean()
spending_By_AgeGroup
```

```python
spending_By_AgeGroup.plot(kind='bar', figsize=(14,8))
plt.title('Average Spending on Products by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Spending')
plt.show()
```

# Is there any Difference in Spending rate between marital status 

```python
spending_By_MaritalStatus =data.groupby('Marital_Status')[Products].mean()
spending_By_MaritalStatus
```

```python
spending_By_MaritalStatus.plot(kind='bar', figsize=(14,8))
plt.title('Average Spending on Products by Marital Status')
plt.xlabel('Marital_Status')
plt.ylabel('Average Spending')
plt.show()
```

# Is the rate Of Purchasing Online And Offline is different ?

```python
data['Total_Online_Purchase'] = data['NumWebPurchases'] + data['NumCatalogPurchases']
data['Total_Ofline_Purchase'] = data['NumStorePurchases']
Online_VS_Ofline = data[['Total_Online_Purchase','Total_Ofline_Purchase']] .sum()
Online_VS_Ofline
```

```python
plt.figure(figsize=(10,6))
sns.barplot(Online_VS_Ofline,palette="spring")
plt.title('Total Purchases: Online vs Store')
plt.xlabel('Purchase Type')
plt.ylabel('Total Purchases')
```

# is there Is relationship between Online Purchase and Income 

```python
plt.figure(figsize=(10,6))
plt.scatter(data['Income'],data['Total_Online_Purchase'],alpha=0.5)
plt.title('Income vs Online Purchases',weight = 'bold')
plt.xlabel('Income')
plt.ylabel('Total Online Purchases')
```
